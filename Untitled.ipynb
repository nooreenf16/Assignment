```python
import matplotlib.pyplot as plt
%matplotlib inline 
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv(r'C:\Users\noore\Downloads\assignment\data2\data\det_data\Task016_Luna\raw\annotations2.csv')
print(dataset.head())
```

```python
dataset[dataset.isnull().any(axis=1)]
```

```python
nodule_dia = np.array(dataset['Nodule_diameter_mm'])
```

```python
min(nodule_dia), max(nodule_dia)
```

```python
w=0.2
fig = plt.figure(figsize = (140, 120))

plt.hist(nodule_dia, bins = np.arange(min(nodule_dia), max(nodule_dia) + w, w), edgecolor='white')

plt.title('Lung Nodule Size in Patients', color='blue', fontsize=340, family='serif')
plt.xticks(fontsize=100, color='red')
plt.yticks(fontsize=100, color='red')

plt.xlabel('Nodule Diameters (mm)', color='blue', fontsize=140)
plt.ylabel('Frequency', color='blue', fontsize=140)

plt.savefig('output.png')
```
